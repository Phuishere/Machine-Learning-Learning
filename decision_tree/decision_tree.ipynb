```python
import numpy as np
import pandas as pd

from id3_algo import ID3
```

```python
# Get columns
columns = ["price", "maintenance", "doors", "persons", "lug_boot", "safety", "decision"]

# Load CSV (skip header, keep as strings)
data = np.genfromtxt("car_evaluation.csv", delimiter=",", dtype=str, skip_header=1)
df = pd.DataFrame(data, columns=columns)
```

```python
# Get unique dict and cols mapping
unique_dict = {}
col_2_idx = {}
idx_2_col = {}
for i, col in enumerate(df.columns):
    unique_dict[col] = df[col].unique()
    col_2_idx[col] = i
    idx_2_col[i] = col
```

```python
# Get data
data = df.to_numpy()

# Into x and y
x = data[:, :-1]
x_cols = columns[:-1]
y = data[:, -1]
```

```python
# Make an instance of the class ID3
id3 = ID3(
    x = x,
    y = y,
    col_names = x_cols
)

# Train in less than 0.5s
id3.traiá¹‡()
```
